```python
reliance_df.Date
```


The Date column has datatype object lets change it to datetime datatype.

```python
reliance_df['Date'] = pd.to_datetime(reliance_df.Date)
```

```python
reliance_df.Date
```


The date column datatype has changed now its easy to access it. lets make different parts of Date column in different columns.

```python
reliance_df['month'] = pd.DatetimeIndex(reliance_df.Date).month
reliance_df['day'] = pd.DatetimeIndex(reliance_df.Date).day
reliance_df['weekday'] = pd.DatetimeIndex(reliance_df.Date).weekday
reliance_df
```


We have added month, day, weekday columns in dataframe.

# Exploratory Analysis and Visualization
In this, we'll do some visualization using some plots, charts in order to make some exploratory analysis.

Let's begin by importingmatplotlib.pyplot and seaborn.

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
reliance_df_high = reliance_df.sort_values(by= 'High', ascending=False).head(10)
reliance_df_high
```

```python
sns.barplot(reliance_df.month, reliance_df.High);
```


It seems that in the month of september and october the stocks value was high which was 2369.

# 2. What was the lowest close in the year 2020 of the stocks ?

```python
reliance_df_close = reliance_df.sort_values(by='Close', ascending=True).head(10)
reliance_df_close
```

```python
sns.histplot(reliance_df['Close']);
```

# 3. What was the turnover in the month of september ?

```python
# lets extract september month data
reliance_df_september = reliance_df[reliance_df.month == 9]

# data of september month turnover
reliance_df_turnoversept = reliance_df_september['Volume']

# sum of turnover in september
reliance_df_turntotal = reliance_df_turnoversept.sum()
reliance_df_turntotal
```

```python
plt.title('Turnover in September')
plt.plot(reliance_df_turnoversept);
```

# 4. What was the count of trade happened in Reliance Shares 2020 ?

```python
reliance_df_trade = reliance_df.High
reliance_df_tradecount = reliance_df_trade.sum()
reliance_df_tradecount
```

```python
sns.barplot(reliance_df.month, reliance_df.High);
```


It seems the trade during the month of september, Octomber were high as compared to other months.

# 5. What was the volume of trades in each month ?

```python
plt.title('Volume of Trades in Each month')
sns.barplot(reliance_df.month, reliance_df.Volume);
```

# 6. What was the highest opening value of Reliance share ?

```python
reliance_df_open = reliance_df.sort_values(by='Open', ascending=False).head(10)
reliance_df_open
```

So the highest opening value was 2360 which was in the month of September.
